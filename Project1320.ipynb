# Part 2 Step 2

```python
nasa_df['rank'] = np.nan

for index, row in df.iterrows():
    final = nasa_df.loc[(nasa_df["start_datetime"].dt.year == row["start_datetime"].year)\
                                & (nasa_df["start_datetime"].dt.month == row["start_datetime"].month) & \
                                (nasa_df["start_datetime"].dt.day == row["start_datetime"].day)]
    if not final.empty:
        cur = final.index[0]
        nasa_df['rank'][cur] = row['rank']

nasa_df.sort_values(by='rank', ascending=True, inplace=True)
nasa_df.head(50)
```

I matched using the year, month, and date because there is not many flares in the same day. Therefore, there is a good chance of finding the correct match for the give top 50 flare. Because of this, there were no duplicate matches. I did this by iterating through the first database and finding matches where the year, month, and day were the same. If there were any matches, I added the rank.
